```python
from typing import List

class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        
        #Only add open parenthesis if open < n
        #Only add closing parenthesis if closed < open
        #Valid iff open == closed == n

        #From the video is clear, but I only don't fully
        #get why stack.pop() is needed.

        #I'll need to review it through a set of 
        #print statements in the notebook.

        #In general, I think I should always see the video
        #of questions I did not get. Even if I understand
        #the final answer, the video explains the reasoning
        #behind it.

        stack = []
        res = []

        def backtrack(openN, closedN):
            
            if openN == closedN == n:
                res.append("".join(stack))
                print(f"In base case, res is: {res}, openN is {openN} and closedN is {closedN}")
                print("--------")
                return
                #print("Ciao") -> This would never get printed, btw

            if openN < n:
                stack.append("(")
                print(f"In openN < n, before backtrack, stack is {stack}, openN is {openN} and closedN is {closedN}")
                backtrack(openN + 1, closedN)
                print(f"In openN < n, after backtrack, stack is {stack}, openN is {openN} and closedN is {closedN}")
                stack.pop()
                print(f"In openN < n, after backtrack and stack.pop(), stack is {stack}, openN is {openN} and closedN is {closedN}")


            if closedN < openN:
                stack.append(")")
                print(f"In closedN < openN, before backtrack, stack is {stack}, openN is {openN} and closedN is {closedN}")
                backtrack(openN, closedN + 1)
                print(f"In closedN < openN, after backtrack, stack is {stack}, openN is {openN} and closedN is {closedN}")
                stack.pop()
                print(f"In closedN < openN, after backtrack and stack.pop(), stack is {stack}, openN is {openN} and closedN is {closedN}")


        backtrack(0, 0)
        return res
```
