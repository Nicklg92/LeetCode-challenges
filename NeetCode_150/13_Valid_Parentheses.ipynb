```python
#The brute force solution requires to sequentially delete all legit couple, hence checking at the end if the remaining 
#string is empty (then it was a valid string) or not (some parenthesis was in fact unmatched).

#It is however quadratic in the length, as in the worst case you'll need to search through (at most) all the n characters for (at most) n times 
#Imagine: ")[]{}(){}[]....."
```

```python
class Solution:
    def isValid(self, s: str) -> bool:
        while '()' in s or '{}' in s or '[]' in s:
            s = s.replace('()', '')
            s = s.replace('{}', '')
            s = s.replace('[]', '')
        return s == ''
```

```python
#Better solution O(n) also time complexity is using stacks. I will just study it, maybe trying to recode it after by memory:
```

```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        closeToOpen = {")":"(", "]": "[", "}":"{"}

        for c in s:
            if s in closeToOpen:
                if stack and stack[-1] == closeToOpen[c]:
                    stack.pop()
                else:
                    return False
            else:
                stack.append(c)
        return True if not stack else False
```
