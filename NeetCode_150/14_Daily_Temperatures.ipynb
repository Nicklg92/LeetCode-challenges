```python
from typing import List

class Solution:
    def dailyTemperatures(self, temperatures: List[int]) -> List[int]:

        counts = []

        for i in range(len(temperatures)):
            count = 0
            j = i + 1
            while j < len(temperatures) and temperatures[i] >= temperatures[j]:
                count += 1
                j += 1
            if j == len(temperatures):
                counts.append(0)
            elif j == i + 1:
                count = 1
                counts.append(count)
            else:
                counts.append(count + 1)
        return counts

#Solved in 30 minutes and 57 seconds. Very similar to suggested Brute Force, quadratic. More elegenat and efficient with stacks.
#Stacks-based solution below.

#But, as often happens, LeetCode does not accept 
```
