```python
from typing import List

class Solution:
    def isValidSudoku(self, board: List[List[str]]) -> bool:
        
        #The sudoku board is a list of lists.
        #Entries are either integers or dots.
        #A duplicate matters only in terms of
        #integers.

        #Extract the unique values of each row.
        #Check their count.
        #If an entry is an integer and its count
        #is greater than 1, return False.

        #Same logic at the column level and 
        #square leveÃ² (to be arranged)

        #A) Checking duplicates at the row level:

        for i in range(len(board)):
            counts = {}
            for j in range(len(board[i])):
                if board[i][j] not in counts:
                    counts[board[i][j]] = 1
                else:
                    counts[board[i][j]] += 1
            for i in counts:
                if i != "." and counts[i] > 1:
                    return False

        print("Rows are ok")

        #B) Checking duplicates at the column level:

        for i in range(len(board[0])):
            counts = {}
            for j in range(len(board)):
                if board[j][i] not in counts:
                    counts[board[j][i]] = 1
                else:
                    counts[board[j][i]] += 1
            print(counts)
            for i in counts:
                print(i)
                if i != "." and counts[i] > 1:
                    return False

        print("Columns are ok")

        #C) Checking duplicates at the block level.
        #I still need to automatically define the blocks,
        #so that the algorithm can stop timely in case.

        i = 0
        j = 0

        while i < len(board[0]):
            while j < len(board):
                block = []
                for x in board[i][j:j+3]:
                    block.append(x)
                print(board[i], board[i][j:j+3])
                for x in board[i+1][j:j+3]:
                    block.append(x)
                print(board[i+1], board[i+1][j:j+3])
                for x in board[i+2][j:j+3]:
                    block.append(x)
                print(board[i+2], board[i+2][j:j+3])

                counts = {}
                print(block)
                for k in block:
                    if k not in counts:
                        counts[k] = 1
                    else:
                        counts[k] += 1
                for m in counts:
                    if m != "." and counts[m] > 1:
                        return False
                j += 3
            i += 3
        print("And also squares are ok")
        return True


#25 minutes and 17 seconds, need to leave the bar, but need to continue, almost there. 
#47 minutes and 59 seconds, solved.

#And all proposed solutions are O(n**2), so nothing 
#to optimize. 
```
