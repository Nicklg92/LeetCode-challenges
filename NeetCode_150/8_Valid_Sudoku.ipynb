```python
"""
a = [['.', '.', '.', '.', '.', '.', '5', '.', '.'], 
['.', '.', '.', '.', '.', '.', '.', '.', '.'], 
['.', '.', '.', '.', '.', '.', '.', '.', '.'], 
['9', '3', '.', '.', '2', '.', '4', '.', '.'], 
['.', '.', '7', '.', '.', '.', '3', '.', '.'], 
['.', '.', '.', '.', '.', '.', '.', '.', '.'], 
['.', '.', '.', '3', '4', '.', '.', '.', '.'], 
['.', '.', '.', '.', '.', '3', '.', '.', '.'], 
['.', '.', '.', '.', '.', '5', '2', '.', '.']]
"""

#Was not passing. That's because the checks for squares was actuallly incorrect.

#This is the correct version, that also checks for squares correctly. The function to reduce the list of lists in a single list
#is actually pretty useful, and could be used to simplify also rows and columns' checks.

class Solution:
    def isValidSudoku(self, board: List[List[str]]) -> bool:
        
        #The sudoku board is a list of lists.
        #Entries are either integers or dots.
        #A duplicate matters only in terms of
        #integers.

        #Extract the unique values of each row.
        #Check their count.
        #If an entry is an integer and its count
        #is greater than 1, return False.

        #Same logic at the column level and 
        #square leveò (to be arranged)

        #A) Checking duplicates at the row level:
        for i in range(len(board)):
            counts = {}
            for j in range(len(board[i])):
                if board[i][j] not in counts:
                    counts[board[i][j]] = 1
                else:
                    counts[board[i][j]] += 1
            for i in counts:
                if i != "." and counts[i] > 1:
                    return False

        print("Rows are ok")

        #B) Checking duplicates at the column level:

        for i in range(len(board[0])):
            #print(i)
            counts = {}
            for j in range(len(board)):
                if board[j][i] not in counts:
                    counts[board[j][i]] = 1
                else:
                    counts[board[j][i]] += 1
            #print(counts)
            for i in counts:
                if i != "." and counts[i] > 1:
                    return False

        print("Columns are ok")

        #C) Checking duplicates at the block level.
        #I still need to automatically define the blocks,
        #so that the algorithm can stop timely in case.

        #We first put the entire board in a flat list

        flat_board = [x for xs in board for x in xs] #-> very useful and worth keeping in mind.

        print(flat_board, len(flat_board))


        #The first cube is 0,1,2,9,10,11,18,19,20. 
        #Id est, with i=0: i, i+1, i+2, i+9, i+10, i+11, i+18,i+19,i+20
        
        #The second block: i + 3 = 3 and the others accordingly.
        #The third block : i + 6 = 6 and the others accordingly.
        
        #The fourth block: i = 27 and the others accordingly.
        #The fifth block: i = 30 and the others accordingly.
        #The sixth block: i = 33 and the others accordingly.

        #The seventh block: i = 54 and the others accordingly.
        #The eight block: i = 57 and the others accordingly.
        #The ninth block: i = 60 and the others accordingly.

        for i in [0, 3, 6, 27, 30, 33, 54, 57, 60]:
            counts = {}
            tmp = []
            for x in [i+0, i+1, i+2, i+9, i+10, i+11, i+18, i+19, i+20]:
                tmp.append(flat_board[x])
            for y in tmp:
                if y not in counts:
                    counts[y] = 1
                else:
                    counts[y] += 1
            for j in counts:
                if j != "." and counts[j] > 1:
                    return False
        return True
                





    
```
